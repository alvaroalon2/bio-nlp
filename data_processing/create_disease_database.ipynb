```python
import pandas as pd 
import numpy as np
```

```python
def set_term(column1,column2):
    if not pd.isnull(column1):
        return column1
    if not pd.isnull(column2):
        return column2
    else:
        return np.nan
```

```python
def set_term_list(column1,column2):
    if isinstance(column1,list):
        return column1
    if isinstance(column2,list):
        return column2
    else:
        return np.nan
```

```python
def separate_id(column,idx):
    for i in column:
        if idx in i:
            column.remove(i)
            if idx == 'MESH' or idx == 'UMLS_CUI':
                return i.split(':')[-1]
            return i
    return np.nan
```
